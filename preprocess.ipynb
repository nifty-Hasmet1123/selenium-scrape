```python
import pandas as pd
from typing import List

with open("D:\Comparison_scrape\info_text\store_details.txt", "r") as file:
    list_of_stores = file.readlines()

extracted_data = [store.strip() for store in list_of_stores]

columns = [
    "Branding Store",
    "Store",
    "Logo-Supervisor",
    "Sales Group",
    "Sales Rep",
    "Orders",
    "External ID",
    "Order Amount",
    "Status",
    "Storez",
    "Created At",
    "Last Fulfilled At",
    "Items Count",
    "Country", 
    "Min Shopify#",
    "Max Shopify#",
    "Formula(Numeric)" 
]
```

```python
def split_data(datas: list) -> list:
    return [data.split() for data in datas]

def make_dict(datas: List[List[str]], column_names: List[str]) -> dict:
    result = []
    for data in datas:
        dictionary_data = {col_name: value for col_name, value in zip(column_names, data)}
        result.append(dictionary_data)
    
    return result

def process_now(datas: list, column_name) -> dict:
    splitted_data = split_data(datas)

    return make_dict(splitted_data, column_name)

res = process_now(extracted_data, columns)
```

```python
df = pd.DataFrame(res).drop(columns = ["Logo-Supervisor", "Storez"])
df
```

```python
df["Store"] = df["Store"].astype("category")
df["Status"] = df["Status"].astype("category")
df["Country"] = df["Country"].astype("category")
df["Branding Store"] = df["Branding Store"].astype("category")
df["Formula(Numeric)"] = df["Formula(Numeric)"].astype("category")
df["Sales Group"] = df["Sales Group"].astype("category")
df["Sales Rep"] = df["Sales Rep"].astype("category")
df["Items Count"] = df["Items Count"].astype("category")
```

```python
df["Created At"] = df["Created At"].str.split("-").str.join(" ")
df["Last Fulfilled At"] = df["Last Fulfilled At"].str.split("-").str.join(" ")

df["Created At"] = df["Created At"].replace("None", pd.NA)
df["Last Fulfilled At"] = df["Last Fulfilled At"].replace("None", pd.NA)
```
